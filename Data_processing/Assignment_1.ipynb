# Task 2
Your program should perform the following tasks:

### 2.1 Report Total Lines
- Count and report the **total number of data lines** stored in the file.

### 2.2 Report Invalid Lines
- Count and report the **total number of invalid lines** in the file.
- A **valid line** must meet the following criteria:
  1. Contains **26 values separated by commas**.  
  2. The **first value** is the **student ID**, formatted as `"N"` followed by **8 digits** (e.g., `N12345678`).  

### 2.3 Report Errors for Invalid Lines
- For any **invalid line**, print an **error message** indicating the line is not valid.


````python
import re       #regular expression
expected_value = 26  # Define the expected number of values per line

STUDENT_ID_PATTERN = re.compile(r'^N\d{8}$')  # Regex to validate student ID: 'N' followed by 8 digits

def validate_line(line_data, line_number):
    values = line_data.split(',')   # Split the line by comma
    if len(values) != expected_value:   # Check if the number of values is correct
        return False, f"Line {line_number}: Incorrect number of value. Expected {expected_value} , got {len(values)}.  Line content: '{line_data}'"

    student_id_str = values[0]  # Check if the first value (student ID) is valid
    if not STUDENT_ID_PATTERN.match(student_id_str):
        return False, f"Line {line_number}: Invalid student ID format. but got: '{student_id_str}'. Line content: '{line_data}'"

    return True, ""

invalid_lines_count = 0

for i, line in enumerate(lines):
    if not line.strip():
        continue

    is_valid, error_message = validate_line(line, i + 1)
    if not is_valid:
        invalid_lines_count += 1
        print(f"[Error] {error_message}")

print("---------------------------------")
print(f"Total number of lines processed: {len(lines)}")
print(f"Total invalid lines of data: {invalid_lines_count}")
````

# Task 3 - Grading Student Exams

Write a program to **grade multiple-choice exams** for a section.  
The exam consists of **25 questions**. Use the following **answer key**:

```python
answer_key = "B,A,D,D,C,B,D,A,C,C,D,B,A,B,A,C,B,D,A,C,A,A,B,D,D"

Scoring Rules:
	•	+4 points for a correct answer
	•	0 points for a skipped question
	•	-1 point for an incorrect answer


````python
answer_key = "B,A,D,D,C,B,D,A,C,C,D,B,A,B,A,C,B,D,A,C,A,A,B,D,D"
answer_key = answer_key.split(",")
print(answer_key)

# Initialize a dictionary to store student scores
student_scores = {}

# Iterate through each line in the file to grade valid students
for i, line in enumerate(lines):
    if not line.strip(): # Skip empty lines
        continue

    # Validate the line before grading (assuming validate_line is defined in global scope)
    is_valid, _ = validate_line(line, i + 1)

    if is_valid:
        # Split the line into values
        values = line.split(',')
        student_id = values[0] # First element is the student ID
        student_answers = values[1:] # All subsequent elements are answers

        current_score = 0
        # Iterate through student answers and compare with answer_key
        # Assuming len(student_answers) is equal to len(answer_key) for valid lines
        for j in range(len(answer_key)):
            student_ans = student_answers[j].strip()
            correct_ans = answer_key[j].strip()

            if student_ans == correct_ans:
                current_score += 4 # Correct answer
            elif student_ans == '':
                current_score += 0 # Blank answer
            else:
                current_score -= 1 # Incorrect answer

        student_scores[student_id] = current_score
    # Invalid lines are skipped for grading

# Print the scores for valid students
print("\n--- Student Scores ---")
for student_id, score in student_scores.items():
    print(f"Student ID: {student_id}, Score: {score}")


````

## Class Statistics:

### 3.1 Count of students scoring > 80
### 3.2 Average score
### 3.3 Highest score
### 3.4 Lowest score
### 3.5 Score range (Highest - Lowest)
### 3.6 Median score:
- Sort all scores in ascending order
- If odd number of students: median = middle value
- If even number of students: median = average of two middle values

### 3.7 Most frequently skipped questions:
- Return a list sorted by question number, number of skips, skip rate
- If multiple questions have the same number of skips, list all

### 3.8 Most frequently incorrect questions:
-	Return a list sorted by question number, number of incorrect answers, error rate
-	Include all questions with the same number of errors

````python
num_advanced_students = 0

for student_id, score in student_scores.items():
    if score > 80:
        num_advanced_students += 1

print(f"Total number of students with scores greater than 80: {num_advanced_students}")

import numpy as np

scores = list(student_scores.values())

# Calculate average score
average_score = np.mean(scores)

# Calculate maximum score
max_score = np.max(scores)

# Calculate minimum score
min_score = np.min(scores)

# Calculate range of scores
range_score = max_score - min_score

# Calculate median score
median_score = np.median(scores)

num_questions = len(answer_key)
skipped_counts = [0] * num_questions
incorrect_counts = [0] * num_questions

# Iterate through each line to analyze individual question responses
for i, line in enumerate(lines):
    if not line.strip():
        continue

    is_valid, _ = validate_line(line, i + 1)

    if is_valid:
        values = line.split(',')
        student_answers = values[1:]

        for j in range(num_questions):
            if j < len(student_answers): # Ensure student_answers has an entry for this question
                student_ans = student_answers[j].strip()
                correct_ans = answer_key[j].strip()

                if student_ans == '':
                    skipped_counts[j] += 1
                elif student_ans != correct_ans:
                    incorrect_counts[j] += 1

# Find the most skipped question
most_skipped_question_index = skipped_counts.index(max(skipped_counts))
most_skipped_count = max(skipped_counts)

# Find the most incorrectly answered question
most_incorrect_question_index = incorrect_counts.index(max(incorrect_counts))
most_incorrect_count = max(incorrect_counts)

print(f"\n--- Statistical Analysis of Scores ---")
print(f"Average Score: {average_score:.2f}")
print(f"Maximum Score: {max_score}")
print(f"Minimum Score: {min_score}")
print(f"Range of Scores: {range_score}")
print(f"Median Score: {median_score}")

print(f"\n--- Question Analysis ---")
print(f"Most skipped question: Question {most_skipped_question_index + 1} (skipped by {most_skipped_count} students)")
print(f"Most incorrectly answered question: Question {most_incorrect_question_index + 1} (incorrectly answered by {most_incorrect_count} students)")
````

````python
## Task 4: Create Output File with Student Grades

Your program should **generate a "results" file** containing **detailed scores for each student**.

- Each line in the output file should include:  
  **Student ID, comma, score**  

- The output file name should be based on the **original input file name**.  
  - Example: If the input file is `class1.txt`, the output file should be named `class1_grades.txt`.
````

````python
# Read the content of the newly created file to verify
import os

# 1. Extract the base filename from the `filename` variable.
base_filename = os.path.splitext(filename)[0]

# 2. Create a new filename by adding '_grades.txt' to the extracted base name.
output_filename = f"{base_filename}_grades.txt"

# 3. Open this new file in write mode ('w') and save the score.
with open(output_filename, 'w') as outfile:
    for student_id, score in student_scores.items():
        outfile.write(f"{student_id},{score}\n")

print(f"Student scores successfully written to '{output_filename}'")

# Read the content of the newly created file to verify
# This section moved after the file is created
with open(output_filename, 'r') as infile:
    verified_content = infile.read()

print(f"\n--- Content of '{output_filename}' ---")
print(verified_content)
````
